```python
#Create Cartograms

import geopandas as gpd
import matplotlib.pyplot as plt

#Load a sample world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

#Scale countries by population for a simple cartogram
world['scaled'] = world['pop_est'] / 10**6    # Population in millions

#Plot the cartogram

world.plot(figsize=(10, 10), column='scaled', legend=True)
plt.title('Cartogram of the World by Population')
plt.show()
```
