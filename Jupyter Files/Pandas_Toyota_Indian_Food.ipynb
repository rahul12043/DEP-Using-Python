```python
#Q1

import pandas as pd
df = pd.read_csv('indian_food.csv')
print(df.head())
```

```python
#Q2

print("Shape: ",df.shape)
print("Size: ", df.size)
print("DataTypes:\n",df.dtypes)
```

```python
#Q3

number = df.shape[0]-1
print("Number of dishes: ", number)
```

```python
#Q4

'''
1. Inference: We can make from this is that average preparation time for indian dishes is 31 minutes and the cook time is 34 minutes.
2. Maximum prep time and cook time are 500 minutes and 720 minutes respectively.
'''
df.describe()
```

```python
#Q5

print(df.isnull().sum())
df.fillna('NaN',inplace=True)
print(df.isnull().sum())
```
